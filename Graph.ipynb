```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = np.linspace(-4, 4, 1000)
y = np.exp(-x**2)

# Set significance level and calculate p-value
sig_level = 0.05
p_value = sum(y[x <= -1.96]) + sum(y[x >= 1.96])

# Create plot
fig, ax = plt.subplots()
ax.plot(x, y)

# Shade area under curve for p-value
ax.fill_between(x[x <= -1.96], y[x <= -1.96], color='red', alpha=0.3)
ax.fill_between(x[x >= 1.96], y[x >= 1.96], color='red', alpha=0.3)

# Add labels and annotations
ax.set_xlabel('Z-score')
ax.set_ylabel('Probability Density')
ax.set_title('Relationship Between P-Values and Significance Levels')

# Add significance level and p-value annotations
ax.axvline(-1.96, color='black', linestyle='--')
ax.axvline(1.96, color='black', linestyle='--')
ax.annotate(f'Significance Level = {sig_level}', xy=(-1.96, 0.4), xytext=(-2.5, 0.5),
            arrowprops=dict(facecolor='black', arrowstyle='->'))
ax.annotate(f'p-value = {p_value:.4f}', xy=(1.96, 0.4), xytext=(2, 0.5),
            arrowprops=dict(facecolor='black', arrowstyle='->'))

# Show plot
plt.show()
```
