```python
import matplotlib.pyplot as plt

# Define the dataset
dataset = [60, 65, 70, 70, 70, 75, 75, 80, 80, 80, 85, 85, 85, 85, 90, 90, 90, 90, 90, 95]

# Calculate the five number summary
minimum = min(dataset)
maximum = max(dataset)
q1 = dataset[int(len(dataset)*0.25)]
median = dataset[int(len(dataset)*0.5)]
q3 = dataset[int(len(dataset)*0.75)]

# Create the boxplot
plt.boxplot(dataset)

# Add labels and title
plt.title('Boxplot for Exam Scores Dataset')
plt.xlabel('Dataset')
plt.ylabel('Score')
plt.xticks([1], ['Scores'])

# Add horizontal line for median
plt.axhline(median, color='red', linestyle='dashed', linewidth=1)

# Add annotations for the five number summary
plt.annotate('Minimum: {}'.format(minimum), xy=(1.1, minimum), xytext=(1.5, minimum))
plt.annotate('Q1: {}'.format(q1), xy=(1.1, q1), xytext=(1.5, q1))
plt.annotate('Median: {}'.format(median), xy=(1.1, median), xytext=(1.5, median))
plt.annotate('Q3: {}'.format(q3), xy=(1.1, q3), xytext=(1.5, q3))
plt.annotate('Maximum: {}'.format(maximum), xy=(1.1, maximum), xytext=(1.5, maximum))

# Show the plot
plt.show()
```
